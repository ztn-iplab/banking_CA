```python
import pandas as pd

# Load the dataset (make sure it's in the same directory as your notebook)
mouse_df = pd.read_csv("logger_mouseactionlog_modified.csv")

# Show general structure
print("Dataset Shape:", mouse_df.shape)
print("\nColumns:\n", mouse_df.columns)
print("\nMissing Values:\n", mouse_df.isnull().sum())
mouse_df.head()
```

```python
# Drop completely empty rows
mouse_df.dropna(how='all', inplace=True)

# Drop rows with missing essential values (like coordinates or timestamp)
essential_cols = ['id', 'action', 'coordinates', 'button', 'delta', 'distance', 'speed', 'user_id', 'timestamp']
mouse_df.dropna(subset=essential_cols, inplace=True)

# Convert timestamp to datetime
mouse_df['timestamp'] = pd.to_datetime(mouse_df['timestamp'], errors='coerce')

# Drop rows with invalid timestamps
mouse_df.dropna(subset=['timestamp'], inplace=True)

# Show cleaned status
print("\nAfter Cleaning:")
print("Remaining rows:", mouse_df.shape[0])
print("Missing Values:\n", mouse_df.isnull().sum())
mouse_df.head()
```
