```python
import pandas as pd
import matplotlib.pyplot as plt

# Load keystroke data
keystroke_df = pd.read_csv("keystroke_user_features.csv")

# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(keystroke_df['dwell_time_mean'].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Mean Keystroke Dwell Time')
plt.xlabel('Dwell Time (seconds)')
plt.ylabel('Number of Users')
plt.grid(True)
plt.tight_layout()
plt.savefig("figure_keystroke_dwell_time.pdf")  # Save as PDF
plt.show()
```
