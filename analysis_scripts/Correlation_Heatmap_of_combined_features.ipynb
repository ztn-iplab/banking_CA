```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load both datasets
keystroke_df = pd.read_csv("keystroke_user_features.csv")
mouse_df = pd.read_csv("mouse_user_features.csv")

# Merge on user_id
combined_df = pd.merge(keystroke_df, mouse_df, on='user_id', how='inner')

# Select only numeric columns
numeric_cols = combined_df.select_dtypes(include=['float64', 'int64'])

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(numeric_cols.corr(), annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap: Keystroke and Mouse Features")
plt.tight_layout()
plt.savefig("figure_correlation_heatmap.pdf")
plt.show()
```
